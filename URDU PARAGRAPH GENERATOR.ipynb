# Article Generation using N-grams

Dataset can be downloaded from  https://www.kaggle.com/datasets/saurabhshahane/urdu-news-dataset 

```python
import pandas as pd
import random
from collections import Counter
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk
```

```python
df = pd.read_excel('urdu-news-dataset-1M.xlsx') #READING DATASET
```

# CREATING NGRAMS

```python
def ngrams(l1,l2, x): 
    for i in range(len(l1)- x+1):
        l2.append(l1[i:i+x])


def token():
    for row in range(200):    #USING 200 rows for faster processing for now
        txt = df.iloc[row,2]
        if type(txt) == str:
            n1.append(txt)
            txt = txt.split()
            ngrams(txt,n2,2)
            ngrams(txt,n3,3)
            ngrams(txt,n4,4)
            ngrams(txt,n5,5)
n1 = []
n2 = []
n3 = []
n4 = []
n5 = []
token()
n1 = ' '.join(n1)
n1 = n1.split()
```

# URDU TEXT GENERATION

```python
def start(l):    
    temp = []
    for i in range(len(n2)):
        if n2[i][0] == l[-1]:
            temp.append(n2[i][1])

    temp = Counter(temp)
    temp = temp.most_common(5)
    if len(temp) != 0:
        x = random.randint(0,len(temp)-1)
        l.append(temp[x][0])
    else:
        x = random.randint(0,len(n2)-1)
        l.append(n2[x][1])
def para1(x):
    n1para = []
    x = x.split()
    for i in range(len(x)):
        n1para.append(x[i])
    while(len(n1para) < 200):
        pos = random.randint(0,len(n1)-1)
        n1para.append(n1[pos])
    n1para = ' '.join(n1para)
    return n1para
def para2(x):
    n2para = []
    x = x.split()
    for i in range(len(x)):
        n2para.append(x[i])
        
    while(len(n2para) < 200):
        temp = []
        for i in range(len(n2)):
            if n2[i][0] == n2para[-1]:
                temp.append(n2[i][1])        
        temp = Counter(temp)
        temp = temp.most_common(5)
        x = random.randint(0,len(temp)-1)
        n2para.append(temp[x][0])
    return ' '.join(n2para)
def para3(x):
    n3para = []
    x = x.split()
    for i in range(len(x)):
        n3para.append(x[i])
    while(len(n3para) < 3):
        start(n3para)

    while(len(n3para) < 200):
        temp = []
        for i in range(len(n3)):
            if (n3[i][1] == n3para[-1]) and (n3[i][0] == n3para[-2]):
                temp.append(n3[i][2])
                
        temp = Counter(temp)
        temp = temp.most_common(5)
        if len(temp) == 0:
            x = random.randint(0,len(n3)-1)
            n3para.append(n3[x][1])
            n3para.append(n3[x][2])
        else:
            x = random.randint(0,len(temp)-1)
            n3para.append(temp[x][0])   
        
    
    return ' '.join(n3para)
def para4(x):
    n4para = []
    x = x.split()
    for i in range(len(x)):
        n4para.append(x[i])
    while(len(n4para) < 4):
        start(n4para)
        
    while(len(n4para) < 200):
        temp = []
        for i in range(len(n4)):
            if (n4para[-3] == n4[i][0]) and (n4para[-2] == n4[i][1]) and (n4para[-1] == n4[i][2]):
                temp.append(n4[i][3])
                
        temp = Counter(temp)
        temp = temp.most_common(5)
        if len(temp) == 0:
            x = random.randint(0,len(n3)-1)
            n4para.append(n4[x][0])
            n4para.append(n4[x][1])
            n4para.append(n4[x][2])
        else:
            x = random.randint(0,len(temp)-1)
            n4para.append(temp[x][0]) 
    return ' '.join(n4para)
def para5(x):
    n5para = []
    x = x.split()
    for i in range(len(x)):
        n5para.append(x[i])
    while(len(n5para) < 5):
        start(n5para)
    
    while(len(n5para) < 200):
        temp = []
        for i in range(len(n5)):
            if (n5para[-4] == n5[i][0]) and (n5para[-3] == n5[i][1]) and (n5para[-2] == n5[i][2]) and (n5para[-1] == n5[i][3]):
                temp.append(n5[i][4])
                
        temp = Counter(temp)
        temp = temp.most_common(5)
        if len(temp) == 0:
            x = random.randint(0,len(n5)-1)
            n5para.append(n5[x][0])
            n5para.append(n5[x][1])
            n5para.append(n5[x][2])
            n5para.append(n5[x][3])
        else:
            x = random.randint(0,len(temp)-1)
            n5para.append(temp[x][0]) 
    return ' '.join(n5para)
inp = 'پیاز' #input here
print(para1(inp))
print('-------------------------------------------------------------------------------------------------------------------------------')
print(para2(inp))
print('-------------------------------------------------------------------------------------------------------------------------------')
print(para3(inp))
print('-------------------------------------------------------------------------------------------------------------------------------')
print(para4(inp))
print('-------------------------------------------------------------------------------------------------------------------------------')
print(para5(inp))
```
